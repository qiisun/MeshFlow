```python
# reading a mesh dataset
import os
import numpy as np
import trimesh
import matplotlib.pyplot as plt

mesh_dataset_dir = "./downloaded_data/shapenet-03636649/objaverse_occ_v5_ids"

mesh_dataset_dir_filtered = "./downloaded_data/shapenet-03636649/objaverse_occ_v5_ids_filter_128"

min_lengths1 = []
for mesh_path in os.listdir(mesh_dataset_dir):
    if mesh_path.endswith(".npz"):
        mesh_data = np.load(os.path.join(mesh_dataset_dir, mesh_path))
        min_length = mesh_data["global_min_length"]
        min_lengths1.append(min_length)

min_lengths2 = []
for mesh_path in os.listdir(mesh_dataset_dir_filtered):
    if mesh_path.endswith(".npz"):
        mesh_data = np.load(os.path.join(mesh_dataset_dir_filtered, mesh_path))
        min_length = mesh_data["global_min_length"]
        min_lengths2.append(min_length)

plt.figure(figsize=(8, 6))
plt.hist(min_lengths1, bins=100, density=True, alpha=0.5, label="original", log=True, range=(0, 0.01))
plt.hist(min_lengths2, bins=100, density=True, alpha=0.5, label="filtered", log=True, range=(0, 0.01))
plt.xlabel("min_length")
plt.ylabel("density")
plt.legend()
plt.title("Histogram of min_length (log y)")

# 多少小于1e-3
# print(f"{np.sum(np.array(min_lengths) < 1e-3) / len(min_lengths)}")
```
